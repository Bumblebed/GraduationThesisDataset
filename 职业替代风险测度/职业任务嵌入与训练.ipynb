激活函数

```python
#sigmoid激活函数
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
#ReLU激活函数
def relu(x):
  return np.maximum(0, x)
#LeakyReLU激活函数
def leaky_relu(x, alpha=0.01):
  return np.where(x >= 0, x, alpha * x)
#Tanh激活函数
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
logit=LogisticRegression(**best_params_)
logit.fit(X_train, y_train)

coefs=logit.coef_[0]
intercept=logit.intercept_

def multi(series):
  result=np.dot(series,coefs)+intercept
  return result

lexical_train_pred_logit=X_train.apply(multi,axis=1)
lexical_train_pred_logit=[sigmoid(y)[0] for y in lexical_train_pred_logit]
meta_data['lexical_train_pred_logit']=lexical_train_pred_logit

lexical_test_pred_logit=X_test.apply(multi,axis=1)
lexical_test_pred_logit=[sigmoid(y)[0] for y in lexical_test_pred_logit]
```

```python
#SVR-------------循环测试验证集上的最优参数
best_params={'C':0,'gamma':0}
best_score=0
for C in np.linspace(1,9.9,100):
  for gamma in np.linspace(0.01,1,100):
    svr=SVR(kernel='rbf',C=C,gamma=gamma)
    svr.fit(X_train,y_train)
    val_score=svr.score(X_val,y_val)
    if val_score>best_score:
      best_score=val_score
      best_params['C']=C
      best_params['gamma']=gamma
print(best_score)
print(best_params)
svr=SVR(C=best_params['C'],gamma=best_params['gamma'])
svr.fit(X_train,y_train)
print(svr.score(X_train,y_train))
```
